```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
kings_data= pd.read_csv("Resources/got.csv")
kings_data
```

```python
# Get attacker and defender 
attacker_data = kings_data["attacker_king"].value_counts()
defender_data = kings_data["defender_king"].value_counts()
```

```python
# Get total battle data
battle_data = attacker_data.add(defender_data, fill_value=0)
```

```python
# Configure plot and ticks
battle_data.plot(kind="bar", facecolor="b
                 ")
```

```python
# Set textual properties
# Set textual properties
plt.title("The Bloodthirst of Kings")
plt.ylabel("Number of Battles Participated In")
plt.xlabel("King")
```

```python
# Show plot
plt.show()
```
